```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv('Mall_Customers.csv')

# Selecting features for clustering
features = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
X = data[features]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 2: Perform K-means clustering
# Determine the optimal number of clusters using the Elbow Method
inertia = []
range_n_clusters = range(1, 11)
for k in range_n_clusters:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
```

```python
# Elbow plot
plt.figure(figsize=(10, 6))
plt.plot(range_n_clusters, inertia, marker='o', linestyle='--')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method to Determine Optimal Clusters')
plt.show()

# Based on the elbow plot, choose an appropriate number of clusters (e.g., 5)
optimal_clusters = 5
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
data['Cluster'] = kmeans.fit_predict(X_scaled)

# Step 3: Visualize the clusters
# Cluster centers
cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)
print("\nCluster Centers:\n", pd.DataFrame(cluster_centers, columns=features))
```

```python
# Pairplot for cluster visualization
sns.pairplot(data, hue='Cluster', vars=features, palette='viridis')
plt.title('Cluster Visualization')
plt.show()

# Distribution of customers across clusters
plt.figure(figsize=(8, 5))
sns.countplot(x='Cluster', data=data, palette='viridis')
plt.title('Number of Customers in Each Cluster')
plt.xlabel('Cluster')
plt.ylabel('Number of Customers')
plt.show()

# Save the clustered data
data.to_csv('clustered_customers.csv', index=False)
```
